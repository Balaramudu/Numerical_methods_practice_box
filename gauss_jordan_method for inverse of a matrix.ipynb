## Gauss jordan method for finding the inverse of a matrix

```python
A_I= [[1, 2 ,3,1,0,0],[2, 3, 2,0,1,0] ,[3,2,6,0,0,1]]
```

```python
A_I
```

```python

def print_matrix(Ab,n):
    print(" ---- ---- ---- ")
    for i1 in range(0,n):
        for i2 in range(0,n+n):
            print(Ab[i1][i2],end = " ")
        print("\n")
```

```python
def gauss_jordan_method_for_inverse(A_I,n):
    for i in range(0,n):
        if (A_I[i][i] ==0 ):
            print("zero division error, try  to swap the rows... ");
            break;
            
        A_Iii = A_I[i][i]
        
        if (A_Iii != 1):
            for j in range(0,2*n):
                A_I[i][j] /= A_Iii
                print_matrix(A_I,n)

        for k in range(0,n):
            if (k!=i):
                a = (-A_I[k][i])
                if (a!=0):
                    for k1 in range(0,2*n):
                        A_I[k][k1] += a*A_I[i][k1]
                        print_matrix(A_I,n)

    return A_I
```

```python
A_I1 = gauss_jordan_method_for_inverse(A_I,3)
```

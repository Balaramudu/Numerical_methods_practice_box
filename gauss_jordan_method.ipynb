## Gauss jordan method to solve system of linear equations without the back_substitution

```python
Ab = [[8,5,11,6],[8,5,8,9],[2,2,5,7]]
#Ab = [[1, 2 ,3,10],[2, 3, 2,6] ,[3,2,6,7]]
```

```python

def print_matrix(Ab,n):
    print(" ---- ---- ---- ")
    for i1 in range(0,n):
        for i2 in range(0,n+1):
            print(Ab[i1][i2],end = " ")
        print("\n")
```

```python
def gauss_jordan_method(Ab,n):
    for i in range(0,n):
        b = 1 # to check for divide by zero error with Abii
        Abii = Ab[i][i] 
        #Ab[i] = Ab[i]/Ab[i][i]
        for j in range(0, n+1):
                
            if(Abii ==0):
                print("\n Some problem with divide by zero, so need to implement bottom row exchange operation ")
                b = 0; break;
            Ab[i][j]  = (Ab[i][j]/Abii)
            print_matrix(Ab,n)
        
        
        if (b==0): break;
            
        for k in range(0,n):
            if (k != i):
                a = (-Ab[k][i])
                for k1 in range(0,n+1):
                    Ab[k][k1] = Ab[k][k1]  + a*(Ab[i][k1])
                    print_matrix(Ab, n)
    return Ab


                
    
```

```python
Ab1 = gauss_jordan_method(Ab,3)
```
